# DeepGlobe EDA

Run **after** `python scripts/prepare_tiles.py`.

1. Class-pixel distribution (per split)
2. Visualize sample tiles + masks
3. Compare 3x3 vs 4x4 tiling choices

```python
import sys, os, json
sys.path.insert(0, '..')
import numpy as np
import cv2
import matplotlib.pyplot as plt

from src.utils import CLASS_NAMES, CLASS_COLORS, label_to_rgb, rgb_to_label

TILES = '../data/tiles'
RAW   = '../data/raw/train'
```

```python
# 1. class distribution
!python ../scripts/class_distribution.py --tiles-root ../data/tiles --split train --out ../outputs/class_dist.json
d = json.load(open('../outputs/class_dist.json'))
names = list(d['fractions'].keys())
vals  = [d['fractions'][n] * 100 for n in names]
plt.figure(figsize=(8, 4))
plt.bar(names, vals)
plt.ylabel('% of pixels (train split)')
plt.xticks(rotation=30)
plt.title('DeepGlobe class distribution')
plt.tight_layout(); plt.show()
```

```python
# 2. sample visualization
splits = json.load(open(os.path.join(TILES, 'splits.json')))
iid = splits['train'][0]
img = cv2.cvtColor(cv2.imread(f'{RAW}/{iid}_sat.jpg'), cv2.COLOR_BGR2RGB)
mask_rgb = cv2.cvtColor(cv2.imread(f'{RAW}/{iid}_mask.png'), cv2.COLOR_BGR2RGB)
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(img); ax[0].set_title(f'{iid} sat'); ax[0].axis('off')
ax[1].imshow(mask_rgb); ax[1].set_title(f'{iid} mask'); ax[1].axis('off')
plt.show()
```
